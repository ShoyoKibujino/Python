```python
# ЗАДАНИЕ 1.

def check_balance(text):
    """
    Проверяет сбалансированность скобок в строке
    """
    stack = []  # используем список как стек
    brackets = {')': '(', '}': '{', ']': '['}

    for char in text:
        # Если открывающая скобка - добавляем в стек
        if char in '({[':
            stack.append(char)

        # Если закрывающая скобка
        elif char in ')}]':
            # Если стек пуст - ошибка
            if not stack:
                return False

            # Проверяем соответствие
            if stack.pop() != brackets[char]:
                return False

    # В конце стек должен быть пустым
    return len(stack) == 0
```

```python
#ЗАДАНИЕ 2.
def calc_rpn(expression):
    stack = []

    for token in expression.split():
        if token == '+':
            b = stack.pop()
            a = stack.pop()
            stack.append(a + b)
        elif token == '-':
            b = stack.pop()
            a = stack.pop()
            stack.append(a - b)
        elif token == '*':
            b = stack.pop()
            a = stack.pop()
            stack.append(a * b)
        elif token == '/':
            b = stack.pop()
            a = stack.pop()
            stack.append(a / b)
        else:
            stack.append(float(token))

    return stack[0]

# Пример
result = calc_rpn("3 4 + 2 *")
print(result)
```
