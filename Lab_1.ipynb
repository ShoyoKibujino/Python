```python
# Лабораторная работа 1.
# Автор: Котова Е.А.
# Группа: ЦИБ-251
# Номер задания: 1.3

# Условие: Дана электрическая цепь, состоящая из 2-х последовательно соединенных проводников (сопротивление каждого известно). Найти общее сопротивление цепи (округление результата необходимо выполнить до 1-го знака после запятой).
# Решение (начало и общий ход мыслей):
# Аналитическое решение.
# Проводники соединены последовательно, поэтому общее сопротивление цепи будет равно сумме их сопротивлений. Сопротивление может быть вещественным числом, поэтому для его хранения необходимо выбрать тип float.

# Код:
r1 = float(input())
r2 = float(input())
total = round(r1 + r2, 1)
print(total)
```
